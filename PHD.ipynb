***Word frequency in captions***

Frecventa cuvintelor in caption-uri

```python
import pandas as pd
import string
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import csv
nltk.download('punkt')
nltk.download('stopwords')

# Read CSV file containing captions
data = pd.read_csv('/content/italia_negative.csv', encoding='latin1')

# Convert non-string values in 'caption' column to string
data['caption'] = data['caption'].astype(str)

# Concatenate all captions into a single string
all_captions = ' '.join(data['caption'])

# Text preprocessing: Convert to lowercase and remove punctuation
all_captions = all_captions.lower()
all_captions = all_captions.translate(str.maketrans('', '', string.punctuation))

# Tokenization
tokens = word_tokenize(all_captions)

# Get English stop words from NLTK
stop_words = set(stopwords.words('english'))

# Additional stop words to remove
additional_stop_words = ["to", "the", "and", "a", "in", "of", "for", "we", "this", "is", "are", "i", "with", "you", "by"]

# Combine both sets of stop words
stop_words.update(additional_stop_words)

# Filter out stop words
filtered_tokens = [word for word in tokens if word not in stop_words]

# Calculate word frequency
word_freq = Counter(filtered_tokens)

# Sort the word frequencies
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Save to CSV file
output_file = 'italia_negative_frequencies.csv'
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Word', 'Frequency'])
    for word, frequency in sorted_word_freq:
        csv_writer.writerow([word, frequency])

print(f"Word frequencies saved to '{output_file}'")
```


**Emotion classification **

Clasificarea emotiilor - neinclus in teza de doctorat


- https://thecleverprogrammer.com/2023/02/06/text-emotions-classification-using-python/

```python
import pandas as pd
import numpy as np
import keras
import tensorflow
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense


data = pd.read_csv("train.txt", sep=';')
data.columns = ["Text", "Emotions"]
print(data.head())
```

```python
texts = data["Text"].tolist()
labels = data["Emotions"].tolist()

# Tokenize the text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
```

```python
sequences = tokenizer.texts_to_sequences(texts)
max_length = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_length)
```

```python
# Encode the string labels to integers
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)
```

```python
# One-hot encode the labels
one_hot_labels = keras.utils.to_categorical(labels)
```

```python
# Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(padded_sequences,
                                                one_hot_labels,
                                                test_size=0.2)
```

```python
# Define the model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1,
                    output_dim=128, input_length=max_length))
model.add(Flatten())
model.add(Dense(units=128, activation="relu"))
model.add(Dense(units=len(one_hot_labels[0]), activation="softmax"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.fit(xtrain, ytrain, epochs=10, batch_size=32, validation_data=(xtest, ytest))
```

```python
# input_text = "She didn't come today because she lost her wallet yestertay!"
input_file_path = '/content/alluvioneemiliaromagna_sentiment_model_italian.xlsx'
df = pd.read_excel(input_file_path)

# Initialize an empty list to store predicted labels
predicted_labels = []
mapped_emotions = []

# Iterate through each caption in the DataFrame
for caption in df['caption']:
    # Preprocess the input text
    input_sequence = tokenizer.texts_to_sequences([caption])
    padded_input_sequence = pad_sequences(input_sequence, maxlen=max_length)

    # Make prediction using the model
    prediction = model.predict(padded_input_sequence)

    # Decode the predicted label using your label encoder
    predicted_label = np.argmax(prediction[0])  # Assuming your model returns a single prediction

     # Map the predicted label to the corresponding emotion name
    predicted_emotion = label_encoder.inverse_transform([predicted_label])

    # Append the predicted label and mapped emotion to the lists
    predicted_labels.append(predicted_label)
    mapped_emotions.append(predicted_emotion[0])

# Add 'emotion' column to the DataFrame
df['emotion_label'] = predicted_labels
df['mapped_emotion'] = mapped_emotions

# Convert 'id' column to string
df['id'] = df['id'].astype(str)

# Save the DataFrame to an Excel file
output_excel_file = 'alluvioneemiliaromagna_emotions.xlsx'
df.to_excel(output_excel_file, index=False)

print(f"Emotion predictions saved to {output_excel_file}")
```

***Color analysis - RGB***

neinclus in teza de doctorat





```python
import cv2
import numpy as np
from skimage.feature import greycomatrix, greycoprops
from skimage import color
import os
import csv

# Function to calculate color statistics for an image
def calculate_color_statistics(image_path):
    # Load an image
    image = cv2.imread(image_path)

    # Image analysis for color
    mean_color = np.mean(image, axis=(0, 1))
    std_color = np.std(image, axis=(0, 1))

    # Convert to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Texture analysis using GLCM (Gray-Level Co-occurrence Matrix)
    glcm = greycomatrix(gray_image, [1], [0], symmetric=True, normed=True)
    contrast = greycoprops(glcm, 'contrast')[0, 0]
    homogeneity = greycoprops(glcm, 'homogeneity')[0, 0]

    return mean_color, std_color, contrast, homogeneity

# Path to the directory containing images
images_directory = '/content/turkey'

# Output CSV file to store statistics
output_file = 'image_statistics.csv'

# Header for the CSV file
header = ['Image', 'Mean_Color_B', 'Mean_Color_G', 'Mean_Color_R', 'Std_Color_B', 'Std_Color_G', 'Std_Color_R', 'Contrast', 'Homogeneity']

# Open the CSV file and write the header
with open(output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(header)

    # Iterate through each image in the directory
    for filename in os.listdir(images_directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust based on image file extensions
            image_path = os.path.join(images_directory, filename)

            # Calculate color statistics for the current image
            mean_color, std_color, contrast, homogeneity = calculate_color_statistics(image_path)

            # Extract the file name (without extension) for the CSV
            image_name = os.path.splitext(filename)[0]

            # Write the statistics to the CSV file
            csv_writer.writerow([image_name, *mean_color, *std_color, contrast, homogeneity])

print(f"Image statistics saved to {output_file}")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to calculate and plot histogram for an image
def plot_image_histogram(image_path, subplot_position):
    # Load a color image
    image = cv2.imread(image_path)

    # Convert the color image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate the histogram for the grayscale image
    hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

    # Plot the histogram in a subplot
    plt.subplot(10, 5, subplot_position)
    plt.title(f"Image {subplot_position}")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.plot(hist, color='black')

# Folder path containing images (replace this with your folder path)
folder_path = '/content/turkey/'

# Get all file names within the folder
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Create and save figures containing histograms
num_images = min(len(image_files), 50)  # Number of images per figure
num_figures = len(image_files) // num_images + 1  # Total number of figures

for i in range(num_figures):
    plt.figure(figsize=(15, 20))

    for j in range(num_images):
        index = i * num_images + j
        if index < len(image_files):
            img_path = os.path.join(folder_path, image_files[index])
            plot_image_histogram(img_path, j + 1)

    plt.tight_layout()
    plt.savefig(f'/content/turkey_hist/histograms_combined_turkey_{i + 1}.png', bbox_inches='tight', pad_inches=0.2)
    plt.close()
```

**Colaj top 20 imagini**

```python
from PIL import Image
import os

# Folder containing images
input_folder = 'aus'

# Output collage image
output_collage = 'australia_top_20.jpg'

# Number of rows and columns
num_rows = 5
num_cols = 4

# Set the dimensions of each image slot and the overall collage
slot_width = 200
slot_height = 200
collage_width = num_cols * slot_width
collage_height = num_rows * slot_height

# Create a blank white image as the base for the collage
collage = Image.new('RGB', (collage_width, collage_height), 'white')

# Get a list of all image files in the input folder
image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]

# Iterate through each image file and paste it onto the collage
for i in range(num_rows):
    for j in range(num_cols):
        # Calculate the position for each slot
        x_position = j * slot_width
        y_position = i * slot_height

        # Check if there are more images to paste
        if image_files:
            # Open the next image
            image_file = image_files.pop(0)
            image_path = os.path.join(input_folder, image_file)
            img = Image.open(image_path)

            # Resize image based on aspect ratio
            aspect_ratio = img.width / img.height
            if aspect_ratio > 1:
                # Landscape image, resize width
                new_width = slot_width
                new_height = int(slot_width / aspect_ratio)
            else:
                # Portrait image, resize height
                new_width = int(slot_height * aspect_ratio)
                new_height = slot_height

            img = img.resize((new_width, new_height), Image.ANTIALIAS)

            # Paste the image onto the collage
            collage.paste(img, (x_position, y_position))

# Save the final collage image
collage.save(output_collage)

print(f"Collage saved to {output_collage}")
```

```python
# Import the required modules and classes
from sparknlp.base import DocumentAssembler, Pipeline
from sparknlp.annotator import LanguageDetectorDL
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

# Initialize Spark session with the required package for Excel
spark = SparkSession.builder.appName("LanguageDetection") \
    .config("spark.jars.packages", "org.apache.spark:spark-excel_2.12:0.14.0") \
    .getOrCreate()

# Step 1: Transforms raw texts to `document` annotation
document_assembler = (
    DocumentAssembler()
    .setInputCol("text")
    .setOutputCol("document")
)

# Step 2: Determines the language of the text
languageDetector = (
    LanguageDetectorDL.pretrained()
    .setInputCols("document")
    .setOutputCol("language")
)

nlpPipeline = Pipeline(stages=[document_assembler, languageDetector])

# Read data from Excel file into a Spark DataFrame
excel_file_path = "/content/australianwildfires_sentiment.xlsx"
data = spark.read.format("com.crealytics.spark.excel") \
    .option("header", "true") \
    .load(excel_file_path)

# Apply NLP pipeline to the "caption" column
processed_data = nlpPipeline.fit(data).transform(data.withColumnRenamed("caption", "text"))

# Save the processed data as an Excel file
output_excel_path = "/content/australia_language.xlsx"
processed_data.select("caption", "language.result").write.format("com.crealytics.spark.excel") \
    .option("header", "true").save(output_excel_path)
```
